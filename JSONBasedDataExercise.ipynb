```python
### In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.
```

```python
print(d[0][:])
```

```python
themeCodes = []
themeMain = []

for row in d:
    r = row
    #print(r)
    for row in r:
        if row['name'] != '':
            print(row['code'],row['name'])
            themeMain.append([row['code'],row['name']])
            themeCodes.append([row['code'],row['name']])
        else:
            themeMain.append([row['code'],row['name']])
```

```python
themeMain = pd.DataFrame(themeMain)
```

```python
themeMain.columns = ['themeCode','theme']
```

```python
themeCodes = pd.DataFrame(themeCodes).drop_duplicates()
```

```python
themeCodes.columns = ['themeCode','theme']
```

```python
print(themeCodes)
```

```python
mergeThemes = themeMain.merge(themeCodes,left_on='themeCode',right_on='themeCode', how='inner')
```
